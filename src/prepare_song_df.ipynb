```python
import hdf5_getters
import numpy as np
import pandas as pd
import os
from pandas.api.types import is_numeric_dtype
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/MSS.csv')
```

```python
numeric_cols = ['artist_latitude', 
'artist_longitude', 
'year', 
'artist_hotttnesss', 
'artist_familiarity', 
'duration', 
'key', 
'key_confidence', 
'loudness', 
'mode', 
'mode_confidence', 
'end_of_fade_in', 
'start_of_fade_out', 
'tempo', 
'time_signature', 
'time_signature_confidence',
'track_id']

str_cols = ['artist_id', 
'artist_mbtags', 
'artist_name', 
'release', 
'similar_artists',
'track_id']


df_numeric = df[numeric_cols]
df.set_index('track_id',inplace=True)
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
b_remove = ['artist_id','artist_mbid','artist_location','track_id','artist_location',
            'artist_name']

for col in b_remove:
    df[col] = df[col].transform(lambda x: x[2:-1])

b_arr_remove = ['artist_mbtags','similar_artists', 'tatums_confidence']

f = lambda x: x[2:-1]
for col in b_arr_remove:
    if len(df[col]) > 0:
        df[col] = [f(x) for x in df[col]]
        
df.head()
```
